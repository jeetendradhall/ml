```python
#initialize plot.ly
import plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
%matplotlib inline
```

```python
df_b_cancer = pd.read_csv ('./datasets/breast_cancer.csv')
```

```python
df_b_cancer = df_b_cancer.drop (['id','Unnamed: 32'], axis = 1)
#df_b_cancer = pd.get_dummies (df_b_cancer, 'diagnosis', drop_first = True)
df_b_cancer.head (5)
```

```python
sns.countplot(x='diagnosis',data = df_b_cancer,palette='BrBG')
```

```python
breast_cancer_dataframe = df_b_cancer
#First Plotly chart - Bar chart to see the count of Malignant and Benign in our data
#create data to feed into the plot - x-axis will hold the name of the diagnosis
#and y-axis will have the counts according to the number of matches found in diagnosis column in our data frame
color = ['red','green']
data = [go.Bar(x=['Malignant','Benign'],
y=[breast_cancer_dataframe.loc[breast_cancer_dataframe['diagnosis']=='M'].shape[0],
   breast_cancer_dataframe.loc[breast_cancer_dataframe['diagnosis']=='B'].shape[0]],
   marker=dict(color=color) 
)]
#create the layout of the chart by defining titles for chart, x-axis and y-axis
layout = go.Layout(title='Breast Cancer - Diagnosis',
xaxis=dict(title='Diagnosis'),
yaxis=dict(title='Number of people')
)
#Imbed data and layout into charts figure using Figure function
fig = go.Figure(data=data, layout=layout)
#Use plot function of plotly to visualize the data
py.offline.plot(fig)
```
